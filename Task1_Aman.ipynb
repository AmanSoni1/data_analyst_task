```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tradelog.csv',index_col=[0] )
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Entry Time'] = pd.to_datetime(df['Entry Time'])
df['Exit Time'] = pd.to_datetime(df['Exit Time'])
```

```python
df = df.sort_values(by='Entry Time')
df = df.reset_index(drop=True)
df
```

```python
df['Total_Price'] = df['Exit Price'] - df['Entry Price']
```

```python
price_list = list(df['Total_Price'])
price_list
```

```python
sum2 = 6500
list1 = []
for i in price_list:
    sum2 = sum2 + i
    list1.append(sum2)
df['port_final'] = list1
```

```python
df
```

## 1. Total Trades

```python
total_t = df.shape[0]
total_t
```

## 2. Profitable Trades

```python
prof_df = df[df['Total_Price']>0]
prof_df = prof_df.reset_index(drop = True)
prof_df
```

```python
profit = prof_df.shape[0]
profit
```

### There are 153 Profit Trades

## 3. Loss Trades

```python
loss_df = df[df['Total_Price']<0]
```

```python
loss_df = loss_df.reset_index(drop = True)
loss_df
```

```python
loss = loss_df.shape[0]
loss
```

### There are 96 Loss-Making Trades

## 4 Win rate: Win Rate is the ratio of profitable trades to all trades

```python
winRate = prof_df.shape[0]/df.shape[0] 
print(winRate)
```

### Winrate = 0.6144

## 5 Average Profit per trade

```python
sum1 = prof_df['Total_Price'].sum()
avgProf = sum1/prof_df.shape[0]
avgProf
```

### Average Profit per trade is 76.0954

## 6. Average Loss per trade

```python
sum2 = loss_df['Total_Price'].sum()
avgloss = sum2/loss_df.shape[0]
avgloss
```

### Average Loss per trade is -105.54375

## 7. Risk Reward ratio: Risk Reward Ratio is the ratio of average profit to average loss.

```python
RiskRew = avgProf/avgloss
RiskRew
```

### 8. Expectancy : (Win Rate x Average Profit) - (Loss Rate x Average Loss), where Loss Rate = 1 - (Win Rate)

```python
expec = (winRate * avgProf) - ((1-winRate) * avgloss)
expec
```

## 9. Average ROR per trade:

```python
import statistics

volatality = statistics.stdev(df['Total_Price'])
volatality
```

```python
AvgReturn = df['Total_Price'].sum() / df.shape[0]
AvgReturn
```

```python
riskreturn = 0.05
ROR = (AvgReturn - riskreturn) /volatality
ROR
```

## 10. Sharpe Ratio:

```python
total_return = df['Total_Price'].sum()
total_return
```

```python
Ret_init = 6500 #given
Ret_fin = Ret_init + total_return
Ret_fin
```

```python
rateofret = ((Ret_fin - Ret_init)/Ret_init )
rateofret
```

```python
sharpe = (rateofret - riskreturn)/volatality
sharpe
```

## 14. Calmar Ratio:

 Calmar ratio = ( Compound Annual Return) / (Maximum drawdown)
 
 CAR = [(Ending Value / Beginning Value) ^ (1 / Number of Years)] - 1

```python
car = ((ending / 6500) ** (1/1)) - 1 # for year 2020
car
```

```python
calmer = car / maxi
calmer
```
